```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
rnpath='/home/pietro/CTA/dtrained/MAE/LST_ResNet50_direction_2020-06-03_17-07/models/ang_reso_plt.npz'
vggpath='/home/pietro/CTA/dtrained/MAE/LST_VGG16_direction_2020-06-06_11-36/models/ang_reso_plt.npz'
resnet = np.load(rnpath)
vgg = np.load(vggpath)
```

```python
rn_theta = resnet['sqrttheta268']
rn_bc = resnet['bin_centers']
vgg_theta = vgg['sqrttheta268']
vgg_bc = vgg['bin_centers']
```

```python
fig = plt.figure(figsize=(12,10))
fig.tight_layout()
plt.semilogx(rn_bc, rn_theta, linewidth=3, label='ResNet50', marker='o')
plt.semilogx(vgg_bc, vgg_theta, linewidth=3, label='VGG16', marker='o')
plt.grid(which='major', linestyle='-')
plt.grid(which='minor', linestyle='--')
# plt.legend()
plt.ylabel(r'$\theta_{68}$ [deg]', fontsize=20)
plt.xlabel('$E_{\mathrm{TRUE}}$ [TeV]', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Angular resolution', fontsize=20)
plt.legend(fontsize=20)


#plt.savefig(folder + '/angular_res.eps', format='eps', transparent=False)
```
