```python
az = events["az"]
print(az)
```

```python
fig = plt.figure(figsize=(10,10))
dir = plt.hist2d(alt*180/3.14,az*180/3.14, bins=200)
plt.xlabel("alt", fontsize=16)
plt.ylabel("az", fontsize=16)
```

```python
d_squared = np.power(az,2) + np.power(alt,2)
fig, ax = plt.subplots()
hist = ax.hist(d_squared, bins=50)
```

```python
# LST pointing direction
az_LST = 3.141592654 # rad
alt_LST = 1.221730476 # rad
point = AltAz(alt=alt_LST * u.rad, az=az_LST * u.rad)

# sources directions
src = AltAz(alt=alt * u.rad, az=az * u.rad)
source_direction = src.transform_to(NominalFrame(origin=point))

delta_alt = source_direction.delta_alt.deg
print(delta_alt.shape)
print("Delta Alt mean: ", np.mean(delta_alt))
print("Delta Alt extrema: ", min(delta_alt), max(delta_alt))

delta_az=source_direction.delta_az.deg
print("Delta Az mean: ", np.mean(delta_az))
print("Delta Az extrema: ", min(delta_az), max(delta_az))

fig = plt.figure(figsize=(10,10))
plt.xlabel("delta_alt", fontsize=16)
plt.ylabel("delta_az", fontsize=16)

dir = plt.hist2d(delta_alt,delta_az, bins=200)#, xlabel="delta_alt")
```
